```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, hamming_loss
import torch
from transformers import AutoTokenizer, BertForSequenceClassification

from data_processing import CONDITIONS, LABELS, NUM_CONDITIONS, NUM_LABELS
from train_bert import BertDataset, create_dataloaders, eval_model
```

### Load Data and Model

```python
experimental_data = pd.read_csv('data/experiment_set.csv')
reports = list(experimental_data['selection'])
labels = experimental_data[CONDITIONS].values.tolist()

train_rep, test_rep, train_lab, test_lab = train_test_split(reports, labels, test_size=0.2, random_state=3)

experimental_data.head()
```

```python
# Load trained model
bert_tokenizer = AutoTokenizer.from_pretrained('models/bert_tokenizer_experimental')
bert_model = BertForSequenceClassification.from_pretrained('models/bert_model_experimental')
criterion = torch.nn.CrossEntropyLoss()
device = torch.device("cpu")
```

```python
# Tokenize reports
train_encodings = bert_tokenizer(train_rep, truncation=True, padding=True, max_length=512)
test_encodings = bert_tokenizer(test_rep, truncation=True, padding=True, max_length=512)
```

```python
# Create dataloader
trainloader, testloader = create_dataloaders(train_encodings, train_lab, test_encodings, test_lab)
```

### Evaluate Model

```python
bert_model.eval()
test_labels = []
test_preds = []
with torch.no_grad():
    for batch in testloader:
        input_ids = batch['input_ids'].to(device) # batch_size x tokenized len
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device) # batch_size x conditions

        outputs = bert_model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits.view(-1, 5, 3) # batch_size x conditions x labels

        preds = torch.argmax(logits, dim=2).cpu().numpy() # batch_size x 5
        test_preds.extend(preds)
        test_labels.extend(labels.cpu().numpy())

labels = np.array(test_labels)
preds = np.array(test_preds)
```

```python
labels.shape, preds.shape
```

```python
# Accuracy and Classification Report by Condition
for i, condition in enumerate(CONDITIONS):
    print(f'{condition}:')
    print(f'Test Accuracy: {accuracy_score(labels[:, i], preds[:, i]):.4f}')
    print('Classification Report:')
    print(classification_report(labels[:, i], preds[:, i], target_names=list(LABELS.values())))
    print('-'*80)
    print("\n")
```

```python
# Confusion matrix for each condition
for i, condition in enumerate(CONDITIONS):
    cm = confusion_matrix(labels[:, i], preds[:, i])
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm_normalized, annot=cm, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive', 'Absent'], yticklabels=['Negative', 'Positive', 'Absent'])
    plt.title(f"Confusion Matrix for {condition}")
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
# Look at actual reports
random.seed(3)
selected_idx = random.sample(range(len(test_rep)), 15)

for idx in selected_idx:
    report = test_rep[idx]
    select_label = labels[idx]
    select_preds = preds[idx]

    print(f'Report: {report}\n')

    for i, condition in enumerate(CONDITIONS):
        l = LABELS[select_label[i]]
        p = LABELS[select_preds[i]]

        print(f'     {condition}: Pred = {p}, Actual = {l}')

    print('-'*80)
```
